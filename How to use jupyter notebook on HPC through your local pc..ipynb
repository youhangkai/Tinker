HPC is a super powerful platform and jupyter notebook is a easy-to-use tool for python.

It is just so cool to use jupyter notebook on your local PC. 

And today, here's the tips to realize it.

1. First of all, login in your HPC account.


2. Make sure you have installed miniconda and python, if you haven't done that, go to https://github.com/youhangkai/Tinker/blob/main/Install%20Python%20on%20HPC.ipynb.

3. Install jupyter package in your environment by following code:

    conda isntall jupyter

4. Open jupyter notebook by following command:

    jupyter notebook --no-browser --port=8888 --NotebookApp.token='your_token'
    
    Remember, the 'your_token' can be changed for any token you like, but make sure you remember it.
    
    The port number can also be changed if 8888 is not avaliable.
    
    

5. Open another local terminal or Command Prompt window, run the following command to set up SSH tunneling or port forwarding:

    ssh -N -L 8888:localhost:8888 username@remote_server_ip
    
    where username is your username and remote_server_ip is the IP address of the remote server.
