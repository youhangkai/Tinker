This is mostly caused by inconsistent definition of geographic coordinates of the data. Some data are georeferenced to the center of the element, some to the top right, some to the bottom left, etc. And this may cause cartopy to think that you have no data at the center.

So our approach is to **transform the center of the projection** and move the center to the edge.

*For example, the code that would have defined the projection is as follows*

```python
lon = lon
lat = lat

projection = ccrs.PlateCarree(central_longitude=0)
```

*And We then we transfer the central longitude with 180*

*Remember that the lontidue map shall also be transferred*

```python
lon = lon - 180
lat = lat

projection = ccrs.PlateCarree(central_longitude=180)
```

Then we can plot a map with no white line then!
